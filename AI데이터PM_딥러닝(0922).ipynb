```python
# 형태소 분석하는 라이브러리 깔기
!pip install kiwipiepy
```

```python
# 형태소 분석 함수 정의하기
import kiwipiepy
kiwi = kiwipiepy.Kiwi()
def extract_nouns(text):
    for token in kiwi.tokenize(text):
        if token.tag in {'NNG', 'NNP'}:
            yield token.form
```

```python
list(extract_nouns('샴푸로 머리를 감으니 시원하다'))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(tokenizer=extract_nouns, min_df=10) # 최소 10건의 특허에 나온 단어만
dtm = cv.fit_transform(df.abstract)
```

```python
word_count = pd.DataFrame({
'word': cv.get_feature_names_out(),
'count': dtm.sum(axis=0).flat
})
word_count.sort_values('count', ascending=False).head(20)
```

```python
# 문서단어 행렬에 SVD 적용, 100차원으로 축소
from sklearn.decomposition import TruncatedSVD
svd = TruncatedSVD(n_components=100, random_state=1234)
doc_emb = svd.fit_transform(dtm)
```

```python
import matplotlib.pyplot as plt
plt.plot(svd.explained_variance_)
```

```python
svd = TruncatedSVD(n_components=15, random_state=1234) # 스크리 플롯(scree plot) 알아서 꺾이는 비탈부분 조
doc_emb = svd.fit_transform(dtm)
```

```python
doc_emb[0]  # 문서 임베딩 = 디지털 지문(고유한 느낌을 숫자로 표현한 것)
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_distances
doc_id = 0
sims = cosine_distances(doc_emb[[doc_id]], doc_emb).flatten() # 유사도 계산(총체적으로 봄)
sim_idx = np.argsort(sims)[:10] # 비슷한 순으로 정렬해서 가장 비슷한 것 10개
df.iloc[sim_idx]
```

# 정보 검색

```python
import pandas as pd
df = pd.read_csv('neurips.zip') # CSV 압축되어 있으면 자동으로 풀어서 엶
df.head()
```

검색을 하려면 먼저 토큰화를 해야한다

```python
# 토큰화 함수 짜기
query = {'natural', 'language'}

import re
def tokenize(text):
    text = text.lower() # 소문자로 변환
    return re.findall(r'\w{2,}', text) # 2글자 이상 단어 추출
```

```python
tokenize('It is known that determinining whether a DEC-P...')
```

### 순차검색

```python
%%time
results = []
for row in df.itertuples():
    words = set(tokenize(row.abstract))
    if query < words: # 검색어가 부분집합이면
        results.append(row.Index)
```

## 해시 테이블(hash table)
- 해시 함수(hash function): 데이터를 고정된 크기의 값으로 매핑하는 함수(예:나머지)

공인인증서: 은행 거래 내역 -> 공인인증서 -> 해시값

암호화폐(블록체인)



```python
from collections import defaultdict
index = defaultdict(set)

for row in df.itertuples():
    words = tokenize(row.abstract)
    for word in words:
        index[word].add(row.Index)

index['language']
```

```python
%%time
results = list(index['natural'] & index['language'])
```
