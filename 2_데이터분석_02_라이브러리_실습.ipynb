#### 1.행렬 데이터를 표현할 수 있다.


```python
import numpy as np

x1 = np.array([5, 3, 1, 0.2])  # 하나의 행렬에 하나의 데이터 타입만 가질 수 있다
print(x1)
print( type(x1) )
print(f'차원: {x1.ndim}')
print(f'구조: {x1.shape}')
print()

x2 = np.array([[5], [3], [1], [0.2]])
print(x2)
print( type(x2) )
print(f'차원: {x2.ndim}')
print(f'구조: {x2.shape}')
```

```python
import numpy as np

arr = np.zeros((2,2)) # 2행 2열 행렬에 값을 모두 0으로 채운다.
print(arr)

arr = np.ones((2,3)) # 2행 3열 행렬에 값을 모두 1으로 채운다.
print(arr)

arr = np.full((2,3), 255) # 2행 3열 행렬에 값을 모두 255으로 채운다.
print(arr)

arr = np.eye(3)  # 3행 3열의 주 대각행렬 요소의 값을 1로 채운다.
print(arr)

arr=np.array(range(20)).reshape((4,5)) # 0~19를 4행 5열 행렬로 만든다.
print(arr)

s = np.sum(arr) # arr행렬 원소 모두 합산하기
print(s)

s = np.sum(arr, axis=0) # 세로축 합산
print(s)

s = np.sum(arr, axis=1) # 가로축 합산
print(s)
```

10.Categoricals (범주화)

```python
df = pd.DataFrame({"id":[1,2,3,4,5,6], "raw_grade":['a', 'b', 'b', 'a', 'a', 'e']})
# 가공하지 않은 성적을 범주형 데이터로 변환합니다.
df["grade"] = df["raw_grade"].astype("category")
df["grade"]
# 범주에 더 의미 있는 이름을 붙여주세요 (Series.cat.categories로 할당하는 것이 적합합니다).
df["grade"].cat.categories = ["very good", "good", "very bad"]
# 범주의 순서를 바꾸고 동시에 누락된 범주를 추가합니다 (Series.cat에 속하는 메소드는 기본적으로 새로운 Series를 반환합니다).
df["grade"] = df["grade"].cat.set_categories(["very bad", "bad", "medium", "good", "very good"])
df["grade"]
```

#### 3.막대(bar) 그래프

```python
import matplotlib.pyplot as plt

# 그래프 크기 설정
plt.rcParams['figure.figsize'] = (6, 3)  # (가로,세로) 인치 단위

# X축, Y축 데이터
X = list(range(2005,2015,1))
Y = [15,7,2,10,8,5,14,9,18,8]

plt.bar(X, Y)     # 막대 그래프
plt.grid(False)
plt.title('년도별 그래프')
plt.show()
```

- 가로막대

```python
plt.barh(X, Y)     # 막대 그래프
plt.grid(False)
plt.title('년도별 그래프')
plt.show()
```

- 그래프 색상 지정

```python
#그래프 색상 지정
colors = ['r','g','b','c','m','y','k','w']  # 기본색상  # Hexa코드 or CSS컬러사용
plt.bar(X, Y,color=colors)
plt.show()
```

```python
# plt.bar(X, Y) # 막대 그래프
# plt.barh(X, Y) # 막대(가로) 그래프
Y2 = [i-2 for i in Y]
plt.bar(X, Y, label='값1')     # 파란색
plt.bar(X, Y2, label='값2')    # 주황색

plt.grid(False)   # 그리드 출력 여부
plt.title('년도별 그래프')
plt.xlabel('년도')
plt.ylabel('값')
plt.legend()
plt.show()
```

-  막대 2개 나란히 나타내기

```python
import matplotlib.pyplot as plt
import numpy as np

# X축, Y축 데이터
X = list(range(2005,2015,1))
Y1= [15,7,2,10,8,5,14,9,18,8]
Y2= [y+1 for y in Y1]
lable = X
X = np.arange(len(X))

plt.bar(X-0.2, Y1, width=0.4, label='Y1')     # 막대 그래프
plt.bar(X+0.2, Y2, width=0.4, label='Y2')     # 막대 그래프
plt.xticks(X, lable)

plt.title('년도별 그래프')
plt.grid(False)
plt.legend()
plt.show()
```

#### 4.원 그래프

```python
import matplotlib.pyplot as plt
import pandas as pd

idx = ['1학년','2학년','3학년','4학년']
val = [16,12,7,5]
per = [val[x]/sum(val) *100 for x in range(len(val))]

plt.pie(per, autopct='%1.1f%%',textprops=dict(color="w"))
plt.title("학년별 동아리 회원수")
plt.legend(idx, title='학년', loc="center right",
           bbox_to_anchor=(1, 0, 0.5, 1))

plt.show()
```

#### 5.혼합그래프 : 선 + 막대

```python
# X축, Y축 데이터
X = list(range(2005,2015,1))
Y1= [15,7,2,10,8,5,14,9,18,8]
Y2= [y+1 for y in Y1]

plt.plot(X, Y1, 'o', linestyle='dashed', c='r', label='Y1') # 선 그래프
plt.bar(X, Y2, width=0.4, label='Y2')     # 막대 그래프
plt.xticks(X, lable)

plt.title('년도별 그래프')
plt.grid(False)
plt.legend()
plt.show()
```
